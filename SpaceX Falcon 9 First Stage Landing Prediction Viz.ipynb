```python
# pandas is a software library written for the Python programming language for data manipulation and analysis.
import pandas as pd
#NumPy is a library for the Python programming language, adding support for large, multi-dimensional arrays and matrices, along with a large collection of high-level mathematical functions to operate on these arrays
import numpy as np
# Matplotlib is a plotting library for python and pyplot gives us a MatLab like plotting framework. We will use this in our plotter function to plot data.
import matplotlib.pyplot as plt
#Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics
import seaborn as sns
```

```python
import pandas as pd

URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv"
df = pd.read_csv(URL)
df.head(5)
```

```python
sns.catplot(y="PayloadMass", x="FlightNumber", hue="Class", data=df, aspect = 5)
plt.xlabel("Flight Number",fontsize=20)
plt.ylabel("Pay load Mass (kg)",fontsize=20)
plt.show()
```

```python
# FlightNumber'a göre LaunchSite'ı ve Class'ı görselleştirme
sns.catplot(x='FlightNumber', y='LaunchSite', hue='Class', data=df, aspect=3, kind='strip', palette='Set2')
plt.title('Flight Number vs Launch Site by Launch Outcome')
plt.xlabel('Flight Number')
plt.ylabel('Launch Site')
plt.grid(True)
plt.show()
```

```python
# FlightNumber'a göre LaunchSite'ı ve Class'ı görselleştirme
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='FlightNumber', y='LaunchSite', hue='Class', palette='coolwarm')
plt.title('Flight Number vs Launch Site (Launch Outcome)')
plt.xlabel('Flight Number')
plt.ylabel('Launch Site')
plt.legend(title='Launch Outcome')
plt.grid(True)
plt.show()
```

```python
# Launch Site'a göre Payload'ı görselleştirme
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='PayloadMass', y='LaunchSite', palette='Set3')
plt.title('Payload vs Launch Site')
plt.xlabel('PayloadMass (kg)')
plt.ylabel('Launch Site')
plt.grid(True)
plt.show()
```

```python
# Payload kütlesi ve fırlatma sitesini görselleştirme
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='PayloadMass', y='LaunchSite', hue='Class', palette='coolwarm')
plt.title('Payload Mass vs Launch Site (Launch Outcome)')
plt.xlabel('Payload Mass (kg)')
plt.ylabel('Launch Site')
plt.legend(title='Launch Outcome')
plt.grid(True)
plt.show()
```

```python
# Yörünge türlerine göre başarı oranını hesaplayalım
orbit_success_rate = df.groupby('Orbit')['Class'].mean() * 100

# Çubuk grafiği oluşturalım
plt.figure(figsize=(10, 6))
orbit_success_rate.plot(kind='bar', color='skyblue')
plt.title('Success Rate by Orbit Type')
plt.xlabel('Orbit Type')
plt.ylabel('Success Rate (%)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# FlightNumber ve Orbit türü arasındaki ilişkiyi görselleştirme
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='FlightNumber', y='Orbit', hue='Class', palette='coolwarm')
plt.title('Flight Number vs Orbit Type (Launch Outcome)')
plt.xlabel('Flight Number')
plt.ylabel('Orbit Type')
plt.legend(title='Launch Outcome')
plt.grid(True)
plt.show()
```

```python
# Payload ve Orbit türü arasındaki ilişkiyi görselleştirme
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='PayloadMass', y='Orbit', palette='Set3')
plt.title('Payload Mass vs Orbit Type')
plt.xlabel('Payload Mass (kg)')
plt.ylabel('Orbit Type')
plt.grid(True)
plt.show()
```

```python
Year=[]
def Extract_year():
    for i in df["Date"]:
        Year.append(i.split("-")[0])
    return Year
Extract_year()
df['Date'] = Year
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Yıllara göre başarılı ve başarısız fırlatmaları sayma
yearly_success_count = df.groupby(df['Date'].dt.year)['Class'].sum()
yearly_total_count = df.groupby(df['Date'].dt.year)['Class'].count()
yearly_success_rate = (yearly_success_count / yearly_total_count) * 100

# Çizgi grafiği oluşturma
plt.figure(figsize=(10, 6))
yearly_success_rate.plot(marker='o', color='green', linestyle='-')
plt.title('Launch Success Rate by Year')
plt.xlabel('Year')
plt.ylabel('Success Rate (%)')
plt.grid(True)
plt.show()
```

```python
features = df[['FlightNumber', 'PayloadMass', 'Orbit', 'LaunchSite', 'Flights', 'GridFins', 'Reused', 'Legs', 'LandingPad', 'Block', 'ReusedCount', 'Serial']]
features.head()
```

```python
# One-Hot Encoding uygulama
features_one_hot = pd.get_dummies(features, columns=['Orbit', 'LaunchSite', 'LandingPad', 'Serial'])

# Sonuçları gösterme
features_one_hot.head()
```

```python
# Tüm sayısal sütunları float64 veri türüne dönüştürme
features_one_hot = features_one_hot.astype(float)

# Sonuçları gösterme
features_one_hot.head()
```

```python
features_one_hot.to_csv('dataset_part_3.csv', index=False)
```
